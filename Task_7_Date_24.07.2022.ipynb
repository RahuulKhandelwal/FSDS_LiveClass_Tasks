# Function Definition: Create Table (with total input as prompt from user)

```python
def create_table(db_name, table_name):
    """Function to create a table in MySql Database
        If the Database also not present it will create it automatically 
        and if the database is present it will simply create a table in the called database
        
        it is a dynamic function which will take all inputs from user as a prompt
        including the no. of columns required in table, Header name and Header type"""
    import logging
    import mysql.connector as conn
    mydb=conn.connect(host='localhost', user='root',passwd='Barney@rocks1985')
    cursor=mydb.cursor()
    logging.basicConfig(filename="Task_7.log", level = logging.INFO)
    
    logging.info(f"Function call: MySql Table creation, Table:{table_name} in Database:{db_name}")
    try:
        logging.info(f"Database {db_name} creation started")
        cursor.execute (f"create database {db_name};")
        logging.info(f"success: database {db_name} created")
    except Exception as e:
        pass
        logging.warning(e)
        
    try:
        no_col=int(input(f" Please enter the no. of columns required for table {table_name}: "))
        l=[]
        
        for i in range (1,no_col+1):
            col_name=input(f"please enter header name for column no.{i}: ")
            col_type=input(f"please enter header Type for column no.{i}: ")
            l.append(col_name+' '+col_type)
        logging.info(f"Table Column Header Names and Types are: {l}")
        s=",".join(l)
        query=f"create table {db_name}.{table_name} "+"("+s+")"+";"
        cursor.execute(query)
        return f"Table Name: {table_name} in Database: {db_name} created successfully in MYSQL"
    except Exception as e:
        return e
        
```

# Function Definition: To Load CSV data to MySql Table automatically

```python
def Load_data(db_name, table_name):
    """Load any file in CSV format to Mysql Table
        
        it takes the path of the file as a prompt from user 
        and automatically export it to the called MySql dataBase and MySql Table"""
    import logging
    logging.basicConfig(filename="Task_7.log", level = logging.INFO)
    
    import mysql.connector as conn
    mydb=conn.connect(host='localhost', user='root',passwd='Barney@rocks1985')
    cursor=mydb.cursor()
    
    logging.info(f"Data Loading start in Table:{table_name} of Database:{db_name}")
    
    try:
        cursor.execute('SET sql_mode=" ";')
        path=input("Please enter destination path of the source file")
        cursor.execute(f"delete from {db_name}.{table_name};")
        logging.critical(f"Previous data deleted from Table:{table_name} of Database:{db_name}")
        query=f"""LOAD DATA INFILE '{path}' 
                 INTO TABLE {db_name}.{table_name}
                 FIELDS TERMINATED BY ',' 
                 Enclosed by '"'
                 LINES TERMINATED BY '\n'
                 IGNORE 1 ROWS
               """
        cursor.execute(query)
        mydb.commit()
        logging.info(f"Data successfully loaded in Table: {table_name} of Database: {db_name}")
        return (f"Data successfully loaded in Table: {table_name} of Database: {db_name}")
        
    except Exception as e:
        logging.critical(e)
        return e
        
    
```
